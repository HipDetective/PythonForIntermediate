## story33_클래스 메서드와 static 메서드

### 클래스 변수에 대한 이해

```python
class Simple:
    def __init__(self):
        self.iv = 10        # iv는 인스턴스 변수, 개체별로 존재하는 변수

s = Simple()
print(s.iv)        # 인스턴스 변수는 객체를 통하여 접근을 한다.
```

```python
class Simple:
    cv = 20     # cv는 클래스 변수, 클래스 Simple에 속함
    def __init__(self):
        self.iv = 10

print(Simple.cv)
s = Simple()
print(s.cv)     # 클래스 변수는 객체를 통해서도 접근 가능
print(s.iv)
```

```python
# count_instance.py
class Simple:
    count = 0       # Simple의 클래스 변수
    def __init__(self):
        Simple.count += 1       # 클래스 변수 count 값 1 증가
    
    def get_count(self):
        return Simple.count     # 클래스 변수 count 값 반환

def main():
    s1 = Simple()       # 클래스 변수 count 값은 1
    print(s1.get_count())
    s2 = Simple()       # 클래스 변수 count 값은 2
    print(s2.get_count())
    s3 = Simple()       # 클래스 변수 count 값은 3
    print(s3.get_count())

main()
```

### static 메서드

```python
# static_method.py
class Simple:
    def sm():        # static 메서드
        print('static method!')
    sm = staticmethod(sm)       # sm 메서드를 static 메서드로 만드는 방법

def main():
    Simple.sm()         # static 메서드는 클래스 이름을 통해 호출 가능
    s = Simple()
    s.sm()              # static 메서드는 객체를 통해 호출 가능

main()
```

```python
# count_instance2.py
class Simple:
    count = 0       # Simple의 클래스 변수
    def __init__(self):
        Simple.count += 1       # 클래스 변수 count 값 1 증가
    
    @staticmethod               # 아래의 메서드를 static 메서드로 선언
    def get_count():            # 매개변수로 self가 없는 static 메서드
        return Simple.count     # 클래스 변수 count 값 반환

def main():
    print(Simple.get_count())   # 객체가 없는 상태에서도 객체의 수를 물을 수 있음
    s = Simple()
    print(Simple.get_count())

main()
```

### class 메서드

```python
# class method.py
class Simple:
    num = 5
    
    @staticmethod
    def sm(i):
        print('st~ 5 + {0} = {1}'.format(i, Simple.num + i))
    
    @classmethod
    def cm(cls, i):
        print('cl~ 5 + {0} = {1}'.format(i, Simple.num + i))

def main():
    Simple.sm(3)        # 클래스 이름 기반의 static 메서드 호출
    Simple.cm(3)        # 클래스 이름 기반의 class 메서드 호출
    s = Simple()
    s.sm(4)             # 객체를 대상으로 항 static 메서드 호출
    s.cm(4)             # 객체를 대상으로 항 class 메서드 호출

main()
```

```python
# class method2.py
class Simple:
    count = 0       # 생성된 객체의 수
    def __init__(self):
        Simple.count += 1
    
    @classmethod
    def get_count(cls):
        return cls.count    # cls에 전달되는 것은 Simple 클래스
    
def main():
    print(Simple.get_count())
    s = Simple()
    print(Simple.get_count())

main()
```

```python
# class method3.py
class Natural:
    
    def __init__(self, n):
        self.n = n

    def getn(self):
        return self.n
    
    @classmethod
    def add(cls, n1, n2):
        return cls(n1.getn() + n2.getn())       # Natural 객체 생성 후 반환
    
def main():
    n1 = Natural(1)
    n2 = Natural(2)
    n3 = Natural.add(n1, n2)        # 반환되는 객체를 n3에 저장
    print('{0} + {1} + {2}'.format(n1.getn(), n2.getn(), n3.getn()))

main()
```

### static 메서드보다 class 메서드가 더 어울리는 경우

```python
# date.py
class Date:     # 날짜를 표현한 클래스
    def __init__(self, y, m, d):
        self.y = y
        self.m = m
        self.d = d
        pass

    def show(self):
        print(f'{self.y}, {self.m}, {self.d}')

    @classmethod
    def next_day(cls, today):       # today 다음 날에 대한 객체 생성 및 반환
        return cls(today.y, today.m, today.d + 1)

def main():
    d1 = Date(2025, 4, 5)
    d1.show()
    d2 = Date.next_day(d1)
    d2.show()
    pass

main()
```

### static 메서드보다 class 메서드가 완전 더 어울리는 경우

```python
# international_date.py
class Date:
    def __init__(self, y, m, d):
        self.y = y
        self.m = m
        self.d = d
    
    def show(self):
        print(f'{self.y}, {self.m}, {self.d}')
    
    @classmethod
    def next_day(cls, today):
        return cls(today.y, today.m, today.d + 1)
    
class KDate(Date):
    def show(self):
        print(f'KOR: {self.y}, {self.m}, {self.d}')

class JDate(Date):
    def show(self):
        print(f'JPN: {self.y}, {self.m}, {self.d}')

def main():
    kd1 = KDate(2025, 4, 12)
    kd1.show()
    kd2 = KDate.next_day(kd1)
    kd2.show()

    jd1 = JDate(2027, 5, 19)
    jd1.show()
    jd2 = JDate.next_day(jd1)
    jd2.show()

main()
```

```python
# person.py
class Person:
    def __init__(self, n, a):
        self.name = n   # 이름
        self.age = a    # 나이
    def __str__(self):
        return '{0}: {1}'.format(self.name, self.age)
        pass

def main():
    p = Person('James', 22)     # 이름:James, 나이: 22
    print(p)
    p.age -= 1      # 실수, 의도는 나이 1살 먹음
    print(p)

main()
```

```python
# person2.py
class Person:
    def __init__(self, n, a):
        self.name = n   # 이름
        self.age = a    # 나이
    
    def __str__(self):
        return '{0}: {1}'.format(self.name, self.age)
        pass
    
    def add_age(self, a):
        if (a < 0):
            print('나이 정보 오류')
        else:
            self.age += a

def main():
    p = Person('James', 22)     # 이름:James, 나이: 22
    print(p)
    p.add_age(1)     # 나이 1살 먹음
    print(p)

main()
```

```python
# person_dict.py
class Person:
    def __init__(self, n, a):
        self._name = n   # 이름
        self._age = a    # 나이
    
def main():
    p = Person('James', 22)     # 이름:James, 나이: 22
    print(p.__dict__)

main()
```

```python
# person_dict2.py
class Person:
    def __init__(self, n, a):
        self._name = n   # 이름
        self._age = a    # 나이
    
def main():
    p = Person('James', 22)     # 이름:James, 나이: 22
    print(p.__dict__)
    p.len = 178                 # 객체에 변수 추가
    p.adr = 'Korea'             # 객체에 변수 추가
    print(p.__dict__)

main()
```

```python
# prt_dict.py
class Simple:
    def __init__(self, n, s):
        self._n = n   # 정수
        self._s = s    # 문자열
    
    def __str__(self):
        return '{0}: {1}'.format(self._n, self._s)
        pass


def main():
    sp = Simple(10, 'my')
    print(sp)           # 변경 전
    sp.__dict__['_n'] += 10     # __dict__에 접근해서 값을 변경
    sp.__dict__['_s'] = 'your'     # __dict__에 접근해서 값을 변경
    print(sp)           # 변경 후

main()
```

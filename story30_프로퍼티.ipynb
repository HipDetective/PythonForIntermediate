## 프로퍼티

### 안전하게 접근하기

```python
# natural.py
class Natural:      # 자연수 표현한 클래스, 1 이상만 저장
    def __init__(self, n):
        if (n < 1):
            self.__n = 1
        else:
            self.__n = n

    def getn(self):
        return self.__n
    
    def setn(self, n):
        if (n < 1):
            self.__n = 1
        else:
            self.__n = n

def main():
    n = Natural(-3)
    print(n.getn())
    n.setn(2)
    print(n.getn())

main()
```

```python
# natural2.py
class Natural:      # 자연수 표현한 클래스
    def __init__(self, n):
        self.setn(n)

    def getn(self):
        return self.__n
    
    def setn(self, n):
        if (n < 1):
            self.__n = 1
        else:
            self.__n = n

def main():
    n1 = Natural(1)
    n2 = Natural(2)
    n3 = Natural(3)
    n1.setn(n2.getn() + n3.getn())
    print(n1.getn())

main()
```

```python
# natural3.py
class Natural:      # 자연수 표현한 클래스
    def __init__(self, n):
        self.setn(n)

    def getn(self):
        return self.__n
    
    def setn(self, n):
        if (n < 1):
            self.__n = 1
        else:
            self.__n = n
    
    n = property(getn, setn)        # 프로퍼티 설정

def main():
    n1 = Natural(1)
    n2 = Natural(2)
    n3 = Natural(3)
    n1.n = n2.n + n3.n
    print(n1.n)

main()
```

### property

```python
# natural4.py
class Natural:      # 자연수 표현한 클래스
    def __init__(self, n):
        self.setn(n)

    n = property()      # property 객체 생성

    def getn(self):
        return self.__n
    
    n = n.getter(getn)      # 위의 getn 메서드를 게터로 등록
    
    def setn(self, n):
        if (n < 1):
            self.__n = 1
        else:
            self.__n = n
    
    n = n.setter(setn)      # 위의 setn 메서드를 세터로 등록

def main():
    n1 = Natural(1)
    n2 = Natural(2)
    n3 = Natural(3)
    n1.n = n2.n + n3.n
    print(n1.n)

main()
```

```python
# natural5.py
class Natural:      # 자연수를 표현한 클래스
    def __init__(self, n):
        self.n = n      # 프로퍼티 n을 통해 접급
    
    n = property()      # property 객체 생성

    def pm(self):
        return self.__n

    n = n.getter(pm)        # 위의 pm을 게터로 등록, 등록했으니 pm이란 이름 불필요

    def pm(self, n):
        if (n < 1):
            self.__n = 1
        else:
            self.__n = n
    
    n = n.setter(pm)        # 위의 pm을 세터로 등록

def main():
    n1 = Natural(1)
    n2 = Natural(2)
    n3 = Natural(3)
    n1.n = n2.n + n3.n
    print(n1.n)

main()
```

### 또 다른 방식

```python
# natural6.py
class Natural:      # 자연수를 표현한 클래스
    def __init__(self, n):
        self.n = n      # 프로퍼티 n을 통해 접급
    
    @property
    def n(self):
        return self.__n

    @n.setter
    def n(self, n):
        if (n < 1):
            self.__n = 1
        else:
            self.__n = n

def main():
    n1 = Natural(1)
    n2 = Natural(2)
    n3 = Natural(3)
    n1.n = n2.n + n3.n
    print(n1.n)

main()
```

```python
# point_3d.py
class Point3D:
    def __init__(self, x, y, z):
        self.x = x      # x좌표
        self.y = y      # y좌표
        self.z = z      # z좌표
        pass

    def __str__(self):
        return '({0}, {1}, {2})'.format(self.x, self.y, self.z)     # 좌표 정보 출력
        pass

def main():
    p1 = Point3D(1, 1, 1)
    p2 = Point3D(24, 17, 31)
    print(p1)
    print(p2)

main()
```

```python
# point_slots.py
class Point3D:
    __slots__ = ('x', 'y', 'z')     # 속성을(변수를) x, y, z로 제한

    def __init__(self, x, y, z):
        self.x = x      # x좌표
        self.y = y      # y좌표
        self.z = z      # z좌표
        pass

    def __str__(self):
        return '({0}, {1}, {2})'.format(self.x, self.y, self.z)     # 좌표 정보 출력
        pass

def main():
    p1 = Point3D(1, 1, 1)
    p2 = Point3D(24, 17, 31)
    print(p1)
    print(p2)

main()
```

## __dict__ 있을 때와 __slots__가 있을 때의 속도 차이

```python
# point_3dict.py
class Point3D:
    def __init__(self, x, y, z):
        self.x = x      # x좌표
        self.y = y      # y좌표
        self.z = z      # z좌표
        pass

    def __str__(self):
        return '({0}, {1}, {2})'.format(self.x, self.y, self.z)     # 좌표 정보 출력
        pass

def main():
    p = Point3D(24, 17, 31)
    print(p.x, p.y, p.z)
    print(p.__dict__['x'], p.__dict__['y'], p.__dict__['z'])

main()
```

```python
# speed_dict.py
import timeit       # 시간 측정을 위한 모듈

class Point3D:

    def __init__(self, x, y, z):
        self.x = x      # x좌표
        self.y = y      # y좌표
        self.z = z      # z좌표
        pass

    def __str__(self):
        return '({0}, {1}, {2})'.format(self.x, self.y, self.z)     # 좌표 정보 출력
        pass

def main():
    start = timeit.default_timer()
    p = Point3D(1, 1, 1)

    for i in range(3000):
        for i in range(3000):
            p.x += 1
            p.y += 1
            p.z += 1
    print(p)

    stop = timeit.default_timer()
    print(stop - start)     # 계산에 걸린 속도 계산 및 출력

main()
```

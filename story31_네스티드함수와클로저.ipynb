## story31_네스티드 함수와 클로저

### 함수를 만들어서 반환하는 함수

```python
def maker(m):
    def inner(n):       # 함수 안에서 정의된 함수, nested 함수라 한다.
        return m * n
    return inner        # 위에서 정의한 nested 함수 반환

f1 = maker(2)
f2 = maker(3)

print(f1(7))
print(f2(7))
```

### 클로저(Closure)

### 저장된 위치 확인하기

```python
def maker(m):
    def inner(n):       # 함수 안에서 정의된 함수, nested 함수라 한다.
        return m * n
    return inner        # 위에서 정의한 nested 함수 반환

f1 = maker(101)
f2 = maker(75)

print(f1.__closure__[0].cell_contents)      # 변수 m의 값을 저장해 놓은 위치
print(f2.__closure__[0].cell_contents)      # 변수 m의 값을 저장해 놓은 위치
```

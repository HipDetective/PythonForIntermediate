### 데코레이터에 대한 이해

```python
def smile():        # 웃는 얼굴 출력
    print("^_^")

def confused():     # 혼란스러운 얼굴 출력
    print("@_@")

smile()
confused()
```

```python
def smile():        # 웃는 얼굴 출력
    print("^_^")

def confused():     # 혼란스러운 얼굴 출력
    print("@_@")

def deco(func):     # 데코레이터 함수
    def df():
        print('emoticon')       # 추가된 기능
        func()                  # 원래 기능
        print('emoticon')       # 추가된 기능
    return df                   # 보강된 기능의 함수를 반환

smile = deco(smile)
smile()

confused = deco(confused)
confused()
```

### 전달 인자가 있는 함수 기반의 데코레이터

```python
def adder2(n1, n2):
    return n1 + n2

def adder3(n1, n2, n3):
    return n1 + n2 + n3

print(adder2(3, 4))
print(adder3(3, 5, 7))
```

```python
def adder2(n1, n2):
    return n1 + n2

def adder3(n1, n2, n3):
    return n1 + n2 + n3

def adder_deco(func):       # 데코레이터 함수
    def ad(*args):          # 전달 인자를 튜플로 묶는다.
        print(*args, sep = ' + ', end = ' ')
        print("= {0}".format(func(*args)))
    
    return ad

adder2 = adder_deco(adder2)
adder2(3, 4)
adder3 = adder_deco(adder3)
adder3(1, 2, 3)
```

### @ 기반으로

```python
def deco(func):     # 데코레이터 함수
    def df():
        print('emoticon')       # 추가된 기능
        func()                  # 원래 기능
        print('emoticon')       # 추가된 기능
    return df                   # 보강된 기능의 함수를 반환

@deco
def smile():        # 웃는 얼굴 출력
    print("^_^")

@deco
def confused():     # 혼란스러운 얼굴 출력
    print("@_@")

smile()
confused()
```

```python
# deco_style.py
def adder_deco(func):       # 데코레이터 함수
    def ad(*args):          # 전달 인자를 튜플로 묶는다.
        print(*args, sep = ' + ', end = ' ')
        print("= {0}".format(func(*args)))
    
    return ad

@adder_deco         # 아래 함수를 데코데이터 adder_deco에 통과시켜라.
def adder2(n1, n2):
    return n1 + n2

@adder_deco         # 아래 함수를 데코데이터 adder_deco에 통과시켜라.
def adder3(n1, n2, n3):
    return n1 + n2 + n3

def main():
    adder2(3, 4)
    adder3(3, 5, 7)

main()
```

### 데코레이터 함수 두 번 이상 통과하기

```python
# deco_style2.py
def deco1(func):        # 데코레이터1
    def inner():
        print('deco1')
        func()
    return inner

def deco2(func):        # 데코레이터1
    def inner():
        print('deco2')
        func()
    return inner

@deco1
@deco2
def simple():
    print('simple')

def main():
    simple()

main()
```
